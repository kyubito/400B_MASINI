We are looking at the behavior of the dark matter halo of M33 under the influence of tidial forces due to MW-M31 systeme merging

So we are ploting a density profile for M33's dark matter halo at different time.

```python
# import modules
import numpy as np
import astropy.units as u
from astropy.constants import G

# import plotting modules
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

# my modules
from ReadFile import Read
from CenterOfMass2 import CenterOfMass
from MassProfile import MassProfile
from OrbitCOM import OrbitCOM
```

```python
# Create MassProfile Objects for M33 Galaxy
M33 = MassProfile("M33",0)
```

```python
# Plot the Density Profile for M33 at snapshot 0.
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)

# Create an array to store Radius, from 0-30 kpc
# intervals of 0.5 kpc
R = np.arange(0.1,31,0.5)

#Create a spherical volume
V = ((4*np.pi*(R**3))/3)

# Plot mass enclosed for Halo 
plt.semilogy(R,M33.MassEnclosed(1,R)/V, color='blue',linestyle="--", linewidth=5)

# Add axis labels
plt.xlabel('Radius (kpc)', fontsize=22)
plt.ylabel(r'(Log Density Enclosed (M$_\odot$/V))', fontsize=22)

#set axis limits
#plt.ylim(1e9,1e12)

#adjust tick label font size
label_size = 22
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size


#add figure text
plt.figtext(0.25, 0.83, 'M33 halo Density Profile at snap 0', fontsize=22)

# Save to a file
#ax.set_rasterized(True)
#plt.savefig('MassProfile_M33.eps', rasterized=True, dpi=350)
```
